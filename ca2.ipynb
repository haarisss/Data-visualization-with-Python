```python
# CLEANING DATASET 1
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('causes-of-accidents-by-severity-of-injury-sustained.csv',sep= ',')

print(df.dtypes,"\n\n",df.count(),"\n\n",df.columns,"\n\n",df.info(),"\n\n",df.isna().any(),"\n\n",pd.notnull(df))
print("\n\nThere are",len(df),"rows of data and",len(df.columns),"columns in this dataset.")
k = df.columns[df.isna().any() == True]
print("There are",df.isna().sum().sum(),"null values and they are in the column",k[0])

plt.figure(figsize=(45, 20))
boxplot = df.boxplot(column=['number_of_accidents'])
boxplot.plot()

plt.show()

print("As you can see, there are too many outliers however I will need these outliers to see what causes of accident results in such high number of accidents.")
df = df.dropna()

# df.to_csv(r"C:\Users\Haaris\OneDrive - Singapore Polytechnic\ca2 pdas\cleaned_causes-of-accidents-by-severity-of-injury-sustained.csv",sep=',')
print("This is how the dataframe looks like after dropping the null rows.")
print("There are now",len(df),"rows of data.")
print("I exported this dataframe into a csv as seen in the commented line above so that I can use the cleaned dataset to modify dataframes which will be seen in the other cells below.")
df
```

```python
# GETTING THE TOTAL NUMBER OF ACCIDENTS OF INJURY AND FATAL COMBINED IN DATASET 1
# EXPORTING INTO A NEW CSV
import pandas as pd
import numpy as np
df = pd.read_csv('cleaned_causes-of-accidents-by-severity-of-injury-sustained.csv',sep= ',')
dfobj = {}
for sev in np.unique(df["accident_classification"]):
    dfobj[sev]={}
    for cause in np.unique(df["causes_of_accident"]):
        for year in np.unique(df["year"]):
            if dfobj[sev].get(cause) is None:
                dfobj[sev][cause]=df[np.logical_and(df['year']==year,np.logical_and(df['causes_of_accident']==cause,df['accident_classification']==sev))]['number_of_accidents'].sum()
            else:
                dfobj[sev][cause]=df[np.logical_and(df['year']==year,np.logical_and(df['causes_of_accident']==cause,df['accident_classification']==sev))]['number_of_accidents'].sum()+dfobj[sev][cause]
                
cleaneddf = pd.DataFrame.from_dict(dfobj)
total = []
for index,row in cleaneddf.iterrows():
    total.append(row.sum())
cleaneddf['Total'] = total
cleaneddf.reset_index(inplace=True)
cleaneddf = cleaneddf.rename(columns={"index":"causes_of_accidents"})


# cleaneddf.to_csv(r"C:\Users\Haaris\OneDrive - Singapore Polytechnic\ca2 davi\total_number_of_accidents.csv",sep=',')
print("From the cleaned dataset,I turned the subgroupings of accident_classification into columns and I made a new column for the total no. accidents of fatal and injury cases combined")
print("\nI exported this dataframe into a csv as seen in the commented line below so that I can use this to plot the graph for DAVI")
cleaneddf
```

```python
# EXPLAINING THE MODIFED DATAFRAME ABOVE
import pandas as pd
import numpy as np
df = pd.read_csv('cleaned_causes-of-accidents-by-severity-of-injury-sustained.csv',sep= ',')
dfobj = {}
for sev in np.unique(df["accident_classification"]):
    dfobj[sev]={}
    for cause in np.unique(df["causes_of_accident"]):
        for year in np.unique(df["year"]):
            if dfobj[sev].get(cause) is None:
                dfobj[sev][cause]=df[np.logical_and(df['year']==year,np.logical_and(df['causes_of_accident']==cause,df['accident_classification']==sev))]['number_of_accidents'].sum()
            else:
                dfobj[sev][cause]=df[np.logical_and(df['year']==year,np.logical_and(df['causes_of_accident']==cause,df['accident_classification']==sev))]['number_of_accidents'].sum()+dfobj[sev][cause]
                
cleaneddf = pd.DataFrame.from_dict(dfobj)
total = []
for index,row in cleaneddf.iterrows():
    total.append(row.sum())
cleaneddf['Total'] = total
cleaneddf.reset_index(inplace=True)
cleaneddf = cleaneddf.rename(columns={"index":"causes_of_accidents"})


print("These are all the causes of accidents. There are 20 causes of accidents. I have accumulated the sum of accident cases for each cause every year into just one row. The values you see below are the accumulated number of accident cases from 2012 to 2018. This is why there are now much fewer rows of column compared to the earlier one.")

print('\nFor example, Changing Lane without Due Care has resulted in 3416 injury cases and 33 fatal cases,a total of 3449 accident cases during the years 2012-2018.')
cleaneddf.iloc[1]
```

```python
# GROUPING  CAUSES OF ACCIDENT AND ACCIDENT CLASSIFICATION,REMOVING ALL THE REPEATED CAUSES IN DATASET 1
# EXPORTING INTO A NEW CSV
import pandas as pd

df = pd.read_csv('cleaned_causes-of-accidents-by-severity-of-injury-sustained.csv',sep= ',')

df = df.groupby(['causes_of_accident','accident_classification'])[['number_of_accidents']].sum()
df.reset_index(inplace=True)

# df.to_csv(r"C:\Users\Haaris\OneDrive - Singapore Polytechnic\ca2 davi\causes_and_classification.csv",sep=',')

print("I grouped the cleaned dataset by causes,classification ann the number of accidents. Now it will show each cause of accident twice, one to show the number of fatal accident cases and the other is the number of injury cases caused by the same cause.")
print("\nI exported this dataframe into a csv as seen in the commented line below so that I can use this to plot the graph for DAVI")
df
```

```python
# EXPLAINING THE MODIFIED DATAFRAME ABOVE
import pandas as pd

df = pd.read_csv('cleaned_causes-of-accidents-by-severity-of-injury-sustained.csv',sep= ',')

df = df.groupby(['causes_of_accident','accident_classification'])[['number_of_accidents']].sum()
df.reset_index(inplace=True)

print('It is now very easy to read data.We can see that are 33 fatal and 3416 injury accidents because of drivers changing lane without due care.')
df.loc[:5]
```
